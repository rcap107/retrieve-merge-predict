```python
import polars as pl
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
base_table = "/home/soda/rcappuzz/work/benchmark-join-suggestions/data/source_tables/company-employees-yadl.parquet"
```

```python
df = pd.read_parquet(base_table)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
enc = OneHotEncoder(handle_unknown="infrequent_if_exist")
```

```python
X = df.drop("target", axis=1)
y = df["target"]
X_enc = enc.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_enc, y, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
```
